```python
import numpy as np

def dfs(mat, start):
    n = len(mat)
    st = [] # LIFO queue
    visited = [False for _ in range(n)]

    st.append(start)

    while st:
        curr_v = st.pop()
        visited[curr_v] = True # visit new vertex
        for neigh in range(n):
            if neigh != curr_v and abs(mat[curr_v][neigh]) > 0: # check the neighbours of current vertex
                if visited[neigh] == False:
                    st.append(neigh)

    for i in range(n): # check if all vertecies were visited
        if visited[i] == False:
            return True

    return False

from queue import Queue

def bfs(mat, start):
    n = len(mat)
    queue = Queue() # FIFO Queue
    visited = [False for _ in range(n)]

    queue.put(start)

    while not queue.empty():
        curr_v = queue.get()
        visited[curr_v] = True # visit new vertex
        for neigh in range(n):
            if neigh != curr_v and abs(mat[curr_v][neigh]) > 0: # check the neighbours of current vertex
                if visited[neigh] == False:
                    queue.put(neigh)

    for i in range(n): # check if all vertecies were visited
        if visited[i] == False:
            return True
        
    return False



def is_reducible(mat):
    n = len(mat)

    for vertex in range(n): # check if every vertex can be visited from every possbile vertex
        flag = dfs(mat, vertex)
        if flag == True:
            return True

    return False

def is_reducible_2(mat):
    
    n = len(mat)

    for vertex in range(n): # check if every vertex can be visited from every possbile vertex
        flag = bfs(mat, vertex)
        if flag == True:
            return True
        
    return False
```
